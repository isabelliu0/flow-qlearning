```python
with open("losscurve.txt", "r") as file:
    content = file.read()
    data = np.array([(0 if i == '' else float(i)) for i in content.split("\n")][:-1])
    
    # convolve data
    n = 100
    filter_arr = 1/n * np.ones(n)
    data = np.convolve(data,filter_arr)
```

```python
plt.title("Learning curve, Step vs. Loss")
plt.xlabel("Step")
plt.ylabel("Loss")
plt.plot(data)
```
