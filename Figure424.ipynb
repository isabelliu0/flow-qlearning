```python
with open("losscurve.txt", "r") as file:
    content = file.read()
    data = np.array([(0 if i == '' else float(i)) for i in content.split("\n")][:-1])
    
    # convolve data
    n = 100
    filter_arr = 1/n * np.ones(n)
    convolved_data = np.convolve(data,filter_arr,mode="same")
    
```

```python
plt.title("Learning curve of FQL Agent, Step vs. Loss")
plt.xlabel("Step Number")
plt.ylabel("Total Loss (Distillation + Behavioral Cloning + Critic)")
plt.plot(data,color="purple",label="Total Training Loss")
plt.plot(convolved_data,color="lightgreen",linewidth=2,label="Smoothed over 100 steps")
plt.legend()
plt.grid(True)
```
